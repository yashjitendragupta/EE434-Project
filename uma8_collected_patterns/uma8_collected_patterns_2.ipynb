This notebook was used to verify the contents of the collected raw signals by using
a peak filter to cut out noise. Then the cleaner signal can be plotted for visual 
inspection. The expectation is 15 rectangular-shaped pulses.

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

def plot(data):
    plt.subplot(7, 1, 1)
    for i_channel in range(7):
        plt.subplot(7, 1, i_channel+1)
        plt.plot(data[:, i_channel])
        plt.ylabel(f'{i_channel}')
    plt.xlabel('Sample')
    plt.show()
```

```python
# 700 Hz, single source

data = np.load('raw_1.npy')

# peak filter at 700Hz
b, a = scipy.signal.iirpeak(700, 10, 48000)
data = scipy.signal.filtfilt(b, a, data, axis=0)

# plot
plot(data)
```
