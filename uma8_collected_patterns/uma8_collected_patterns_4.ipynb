```python
import numpy as np
from matplotlib import pyplot as plt

measured_angles = [0, 45, 90, 135, 180]
angle_range = range(180+1)

def get_theoretical_pattern(w):
    c = 343
    X = np.array([ # estimated positions of the microphones
        [0, 0],
        [0, 0.045],
        [0.039, 0.023],
        [0.039, -0.023],
        [0, -0.045],
        [-0.039, -0.023],
        [-0.039, 0.023],
    ])

    target_dir = np.array([1, 0]) # target was about this way (direction of USB port)
    target_k = (2*np.pi*w/c)*target_dir
    target_vector = (1/len(X))*np.exp(1j*X@target_k)

    incoming_dir = [np.array([np.cos(np.deg2rad(a)), np.sin(np.deg2rad(a))]) for a in angle_range]
    incoming_k = [(2*np.pi*w/c)*dir for dir in incoming_dir]
    incoming_vectors = [np.exp(-1j*X@k) for k in incoming_k]

    return np.array([np.abs(incoming_vectors.dot(target_vector)) for incoming_vectors in incoming_vectors])
```

```python
# 700 Hz, single source

measured_array_pattern = np.array(
    [0.888888889, 0.841269841, 0.983333333, 0.905555556, 0.777777778]
)
measured_array_pattern_err = np.array(
    [0.165052263, 0.176915965, 0.103495326, 0.104125836, 0.168122854]
)

# normalize the results using the first angle
first_angle_gain = measured_array_pattern[0]
measured_array_pattern /= first_angle_gain
measured_array_pattern_err /= first_angle_gain

theoretical_array_pattern = get_theoretical_pattern(700)

plt.title('Measured vs theoretical array pattern at 700 Hz')
plt.scatter(measured_angles, measured_array_pattern, label='measured')
plt.errorbar(measured_angles, measured_array_pattern, yerr=measured_array_pattern_err, fmt='none')
plt.plot(angle_range, theoretical_array_pattern, color='darkorange', label='theoretical')
plt.ylim([0, 1.3])
plt.xlabel('Angle (deg)')
plt.ylabel('Gain')
plt.legend()
plt.show()
```
